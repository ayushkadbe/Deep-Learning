```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv(r"C:\Users\AYUSH\Desktop\Keras\Module 3\concrete_data.csv")
concrete_data.head()
```

```python
#lets check datapoints
concrete_data.shape
```

```python
#let's check missing values
concrete_data.describe()
concrete_data.isnull().sum()
```

```python
#Split data into Predictor variables(independent i.e who will be used to predict/test/explain the target variables) and target variables (dependent i.e whose values will be modeled)
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
target = concrete_data['Strength']
```

```python
predictors.head()
target.head()
```

```python
predictors_norm = (predictors - predictors.mean())/predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1]
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
# fit the model
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```
